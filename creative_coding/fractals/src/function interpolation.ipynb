```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
range_size = 4
step_size = 0.1
```

```python
x = np.arange(0, range_size, step_size)
```

```python
f_square = lambda x: x ** 2
f_cube   = lambda x: x ** 3
```

```python
plt.plot(x, f_square(x))
plt.plot(x, f_cube(x))
```

```python
f_inter = lambda x: (f_square(x) + f_cube(x)) / 2
```

```python
plt.plot(x, f_square(x))
plt.plot(x, f_cube(x))
plt.plot(x, f_inter(x))
```

```python
n_interp         = 5
interp_functions = []
for i in range(n_interp):
    new_function = lambda x, i = i, n_interp = n_interp: (
        (n_interp - i) * f_square(x) +
        i * f_cube(x)
    ) / n_interp
    interp_functions.append(new_function)
```

```python
for i in [0, -1, 1, 2, 3]:
    plt.plot(x, interp_functions[i](x))
```
